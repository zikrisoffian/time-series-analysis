## Properties of time series

#### < General rule for time-series analysis >

Before going any further into our analysis, our series has to be made stationary.

Stationary: property of exhibiting constant statistical properties (mean, variance, autocorrelation, etc.). \
If the mean of a time-series increases over time, then it’s not stationary.

#### < Properties and types of series >

Each time series can be thought as a mix between several parts :

- A trend (upward or downwards movement of the curve on the long term)
- A seasonal component
- Residuals

< Trends >

Trends: A long term increase or decrease in the data. This can be seen as a slope (is doesn’t have to be linear) roughly going through the data.

< Seasonality >

Seasonality: A time series is said to be seasonal when it is affected by seasonal factors (hour of day, week, month, year, etc.). Seasonality can be observed with nice cyclical patterns of fixed frequency.


< Residual >

Residual (error term) : They are the difference between an observation and its predicted value at each time step. Residual is unexplained fluntuation in the real value

```python
# import library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# ignore warnings
import warnings
warnings.filterwarnings("ignore")

# statsmodels seasonal_decompose
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# import Airline datasets

url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
df = pd.read_csv(url,index_col=0, header = 0, parse_dates=['Month'])

# plot the Airline dataset grapgh
plt.figure(figsize = (15,8))
plt.plot(df)
```

```python
# plot the decomposition

plt.rc("figure", figsize=(20,10))
result = seasonal_decompose(df, model='multiplicative')
result.plot()
plt.show()
```
